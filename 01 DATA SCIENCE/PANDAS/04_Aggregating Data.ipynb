```python
import pandas as pd
import numpy as np
import os

fp = os.getcwd()

df = pd.read_csv(fp + r"\data\Input\titanic_train.csv") 

# shortening the DataFrame for readability
df = df[["PassengerId", "Survived", "Pclass", "Sex", "Age", "Fare", "Embarked"]]
df.head(3)
```

# Group by aggregation
Group rows together by aggregating their values with a given aggregation function. Basic aggregation functions are (sum, mean, min, max, count or median)

```python
grouped = df.groupby(["Sex", "Survived"]).agg(
                      Median_Age = ("Age", "median"),
                      Count_Id = ("PassengerId", "count")
                      )

grouped
```

# Pivoting
By Pivoting we can achieve the same as grouping, but we can also specify the columns we want to pivot by. If we don't want to aggregate by function, we just pass values without an 'aggfunc' but this only works if the data isn't aggregated so theres only a single datapoint/value for that index/column combination (see example 2).

```python
df.pivot_table(values=["Fare", "Pclass"],
               index=["Survived"], 
               columns=["Embarked"], 
               fill_value=0,
               aggfunc={
                   "Fare": "mean",
                   "Pclass": "mean"
               }
              )
```
