# Describe DataFrame
Methods that give out basic information about the DataFrame. You should always run these methods before taking any actions or doing any analysis on the data, to get an 
overview of the DataFrame. You can see the details about the DataFrame, how many entries are in the DataFrame, see what columns are missing values, what datatype each 
column has, which columns you maybe need to convert etc etc.

### Head
Prints the first n rows

```python
df.head(5)
```

### Tail
prints the last n rows

```python
df.tail(5)
```

### Describe
prints statistically overview (count, mean, min, max) of every column with numerical datatype

```python
df.describe()
```

### Columns
prints out all columns in a list

```python
df.columns
```

### Dtypes
prints out every columns datatypes

```python
df.dtypes
```

### Count
Prints the amount of values in each column

```python
df.count()
```

### Correlation
gives a basic correlation matrix between all numerical columns

```python
df.corr()
```
